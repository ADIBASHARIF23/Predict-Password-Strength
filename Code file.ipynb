```python
#passing the entire dataset to the array named 'password_tupple'
password_tupple= np.array(data)
password_tupple
```

```python
#to make the model more ribust 
import random
random.shuffle(password_tupple)
```

```python
#saperating independent (password) & dependent ( strength) features
x = [labels[0] for labels in password_tupple]
y = [labels[1] for labels in password_tupple]
```

```python
x
```

```python
y
```

**Applying TF-IDF on the data**

```python
#before applying TF-IDF on the data , I need to make sure whether the data is string or not . Cause String type data is all we need 
def word_devide_char(inputs):
    character = []
    for i in inputs:
        character.append(i)
    return character
```

```python
word_devide_char('yj9q3f8p')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorized_data = TfidfVectorizer(tokenizer= word_devide_char)
```

```python
X = vectorized_data.fit_transform(x)
```

```python
X.shape
```

```python
vectorized_data.get_feature_names_out()
```

```python
first_document_vector = X[0]
first_document_vector 
```

```python
#to get the data in vectorized form 
first_document_vector.T.todense()
```

```python
df= pd.DataFrame(data=first_document_vector.T.todense(),index=vectorized_data.get_feature_names_out(),columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending=False)
```

**Applying classification model on data**

```python
#since the dataset is imbalanced ; we can't apply linear regression. 
#applying logistic regression on data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

```python
#predict the model with the test data
y_pred = clf.predict(X_test)
y_pred
```

```python
#predict the model with random data
dt= np.array(['@ggerer#1314324'])
pred = vectorized_data.transform(dt)
clf.predict(pred)
```

**Checking accuracy of model**

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm =  confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
